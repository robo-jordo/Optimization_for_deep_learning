## question 2.3: random search

Implementing random search in Python
Implement the random search algorithm described in Section 0.6 in Python,
and use your implementation to minimize the function

g(w0, w1) = tanh(4w0 + 4w1) + max(0.4w0^2,1)

Search through P = 1000 random directions at each step, α = 1, a maximum
of 8 steps, and an initial point w0 = (2,2)

Here there is an entire region of global
minima exists where g(w1, w2) = 1 the method - as clumsy as it is given the
settings - quickly finds a global minimum. Hint: You can easily visualize this
function since it only takes in two inputs.

```python
# this is needed to compensate for %matplotlib notebook's tendancy to blow up images when plotted inline
%matplotlib notebook
from matplotlib import rcParams
rcParams['figure.autolayout'] = True
```

Import libraries

```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
```

Set constants

```python
P = 1000
alpha = 1
K = 8
omega_k = (2,2)
```

Define the function g

```python
def g(w):
    evaluation = np.tanh(4*w[0] + 4*w[1]) + max(0.4*w[0]*w[0],1)+1
    return evaluation
```

Do the random local minima search

```python
omega_k = (-2,-2)
value = g(omega_k)

x_list_1 = [omega_k[0]]
y_list_1 = [omega_k[1]]
z_list_1 = [value]

for i in range(K):
    for j in range(P):
        random_dir = np.random.randn(2,1)
        normalized_denom = math.sqrt(random_dir[0]**2+random_dir[1]**2)
        x_dir = (random_dir[0]/normalized_denom)**2
        y_dir = (random_dir[1]/normalized_denom)**2
        if (j==0):
            random_point = (omega_k[0]+x_dir*alpha,omega_k[1]+y_dir*alpha)
            best_point =  random_point
            random_value = g(random_point)
        else:
            random_point = (omega_k[0]+x_dir*alpha,omega_k[1]+y_dir*alpha)
            if(g(random_point)<random_value):
                random_value = g(random_point)
                best_point = random_point
                
    if(random_value<value):
        value = random_value
        omega_k = best_point
        x_list_1.append(omega_k[0])
        y_list_1.append(omega_k[1])
        z_list_1.append(value)
        print("step:"+str(i+1)+' : evaluation ='+str(value[0]))
    else:
        print("step:"+str(i+1)+' : evaluation ='+str(value[0]))
```

```python
fig, ax = plt.subplots()
ax.scatter(x_list_1,y_list_1,c='red')
plt.show()
```
