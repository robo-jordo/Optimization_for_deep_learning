Import autograd

```python
# the import statement for matplotlib
import matplotlib.pyplot as plt
import autograd.numpy as np  
```

Create a function to mimic the mathematical function being used

```python
def g(w):
    evaluation = 0.02*(w**4+w**2+10*w)
    return evaluation
```

```python
# create a sample of points to plot over 
w_vals = np.linspace(-5,5,200)
g_vals = []
# evaluate the function over each of these values - one can use an explicit for-loop here instead of a list comprehension
for v in w_vals:
    g_vals.append(g(v))

# plot
fig, ax = plt.subplots(1, 1, figsize=(6,3))
ax.plot(w_vals,g_vals)
plt.show()
```

Create function to mimic the gradient of g

```python
def grad_g(w):
    evaluation = 0.02*(4*(w**3)+2*w+10)
    return(evaluation)
```

```python
# create a sample of points to plot over 
w_vals = np.linspace(-5,5,200)
g_vals = []
# evaluate the function over each of these values - one can use an explicit for-loop here instead of a list comprehension
for v in w_vals:
    g_vals.append(grad_g(v))

# plot
fig, ax = plt.subplots(1, 1, figsize=(6,3))
ax.plot(w_vals,g_vals)
plt.show()
```

Write the gradient descent function

```python
def gradient_descent(alpha,max_its,w_start):
    w = w_start
    weight_history = [w_start]
    cost_history = [g(w_start)]
    for i in range(max_its):
        grad_eval = grad_g(w)
        w = w - alpha*grad_eval
        weight_history.append(w)
        cost_history.append(g(w))
    return cost_history
```

Gradient descent with alpha = 1

```python
gradient_descent(1,1000,2)
```

Gradient descent with alpha = 0.1

```python
costs2 = gradient_descent(0.1,1000,2)
```

Gradient descent with alpha = 0.01

```python
costs3 = gradient_descent(0.01,1000,2)
```
